### 4.2 Qual o jogo desse vídeo game com maior número de vendas nos NA ? 

```python
data["Platform"]
```

```python
selection =  data["Platform"] == "DS"
selection
```

```python
data_ds = data[selection]
data_ds
```

```python
data_ds.sort_values(by=["NA_Sales"], ascending=False)
```

### 4.3 Sabendo que esse jogo lançou em 2006, quais foram os jogos mais vendidos no NA entre 2005 e 2007 ?

```python
selection =  (data["Year"] >= 2005.0) & (data["Year"] <= 2007.0)
selection
```

```python
data_between_2005_and_2007 = data[selection]
data_between_2005_and_2007.sort_values(by=["NA_Sales"], ascending=False).head(20)
```

### 4.4 Levando em consideração o dataframe anterior (20 primeiros por vendas em NA entre 2005 e 2007), qual o valor de vendas (NA) agrupado por publicadora ?

```python
filtered_data = data_between_2005_and_2007.sort_values(by=["NA_Sales"], ascending=False).iloc[0:20]
filtered_data
```

```python
filtered_data.groupby("Publisher").sum()
```

### 4.3 Sabendo que esse jogo lançou em 2006, quais foram os jogos mais vendidos no NA entre 2005 e 2007 ?

```python
selection =  (data["Year"] >= 2005.0) & (data["Year"] <= 2007.0)
selection
```

```python
data_between_2005_and_2007 = data[selection]
data_between_2005_and_2007.sort_values(by=["NA_Sales"], ascending=False).head(20)
```

### 4.4 Levando em consideração o dataframe anterior (20 primeiros por vendas em NA entre 2005 e 2007), qual o valor de vendas (NA) agrupado por publicadora ?

```python
filtered_data = data_between_2005_and_2007.sort_values(by=["NA_Sales"], ascending=False).iloc[0:20]
filtered_data
```

```python
filtered_data.groupby("Publisher").sum()
```

# 5. Tratando base de dados

```python
data.head()
```

### 5.1 - Removendo colunas não tão interessantes

```python
t_data = data.drop(columns=['Rank', 'NA_Sales', "EU_Sales", "JP_Sales", "Other_Sales"])
t_data
```

### 5.2 - Data como um inteiro

```python
t_data["Formatted Year"] = t_data["Year"].fillna(0.0).astype(int)
t_data.head()
```

```python
t_data.shape
```

```python
selection =  t_data["Year"] > 1000
selection
```

```python
t_data = t_data[selection]
t_data.shape
```

```python
t_data = t_data.drop(columns=["Year"])
t_data.head()
```

### 5.3 - Renomear colunas

```python
t_data = t_data.rename(columns = {"Global_Sales": "Global_Sales (MM)", "Formatted Year": "Year"})
t_data.head()
```

### 5.4 - Selecionar apenas os dados com vendas globais maior que 1 MM

```python
t_data.shape
```

```python
selection =  t_data["Global_Sales (MM)"] >= 1.00
selection
```

```python
t_data = t_data[selection]
t_data.shape
```
